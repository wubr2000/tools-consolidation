## Part 3 - Predictive modeling

```python
File = open("uber_data_challenge.json",'r')
Data = File.read()
File.close()
finaljson = json.loads(Data)
```

```python
churn_df = json_normalize(finaljson)
```

```python
churn_df.head()
```

```python
churn_df.uber_black_user = [int(ub) for ub in churn_df.uber_black_user]
#Turn these into datetime objects
churn_df.last_trip_date = pd.to_datetime(churn_df.last_trip_date)
churn_df.signup_date = pd.to_datetime(churn_df.signup_date)
```

```python
#Define "Retained" as someone who is active on their 6th Month on the system
#"Active"/Retained is defined as a user who took a trip in the preceding 30 days
#Therefore, add 6 months to their 'signup_date' and then subtract 30 days to see if they have been active within those 30 days.

from dateutil.relativedelta import relativedelta

churn_df['cutoff_date'] = [signup + relativedelta(month=6) - timedelta(days=30) for signup in churn_df.signup_date]

#We want to find whether a user has been retained/"active" - took a trip in the preceding 30 days (from when the dataset was taken)
churn_df['retained'] = 1*(churn_df.last_trip_date > churn_df.cutoff_date)
```

```python
churn_df.head()
```

### Fractions of Retained Users

```python
print 'Fraction of users retained:', churn_df['retained'].sum() *1.0 / churn_df.shape[0]
```

### Dummify Categorical Variables

```python
#Dummify 'city' and 'phone' variables
#Add dummy columns to matrix and remove original column from matrix
columns_to_dummy = ['city','phone']
for column in columns_to_dummy:
    dummies = pd.get_dummies(churn_df[column], prefix = column)
    churn_df[dummies.columns] = dummies
    churn_df = churn_df.drop(column,1)
```

### Dealing with Missing Values and Features related to Ratings

```python
#Check for missing values
pd.isnull(churn_df).sum()
```

```python
#'avg_rating_by_driver' - some missing but a very small %
#'avg_rating_of_driver' - 16% MISSING! check whether it's systematically missing - one class has more missing than the other
```

```python
#Fill in missing 'avg_rating_by_driver' by using median
churn_df.avg_rating_by_driver = churn_df.avg_rating_by_driver.fillna(churn_df.avg_rating_by_driver.median())
```

```python
print "Fraction of missing values in 'avg_rating_of_driver' who are Retained customers:", \
churn_df.ix[np.where(pd.isnull(churn_df.avg_rating_of_driver))]['retained'].sum()*1.0 / churn_df.ix[np.where(pd.isnull(churn_df.avg_rating_of_driver))].shape[0]
```

```python
##Missing value in "avg_rating_of_driver" is more prevalent amongst customers who are NOT active/retained
##In fact, based on this, it might make sense to create a variable that indicates 
#whether 'avg_rating_of_driver' is missing or not as a feature for the model.
```

```python
x = churn_df.ix[np.where(pd.notnull(churn_df.avg_rating_of_driver))][churn_df.ix[np.where(pd.notnull(churn_df.avg_rating_of_driver))].retained==1]['avg_rating_of_driver']
y = churn_df.ix[np.where(pd.notnull(churn_df.avg_rating_of_driver))][churn_df.ix[np.where(pd.notnull(churn_df.avg_rating_of_driver))].retained==0]['avg_rating_of_driver']

plt.hist(x, bins=10, alpha=0.5, label='Retained', normed=True)
plt.hist(y, bins=10, alpha=0.5, label='Churned', normed=True)

plt.title("Histogram of 'avg_rating_OF_driver' by Churned/Retained Customers (excl. missing values)", fontsize=12)
plt.legend(loc='upper left')
plt.show();
```

```python
#As can be seen, there is very difference between the churned/retained customers in terms of distribution of their ratings of the driver.
#Therefore, we could use the fact that missing value for driver rating as a signal for churn/retained rather than using the rating itself.
churn_df['avg_rating_of_driver_missing'] = np.where(pd.isnull(churn_df.avg_rating_of_driver), 1, 0)
```

### Features related to Dummified Categorical Variables

```python
#Chi-square test of independence for categorical variables with retained/churned label
chi2_list = []
p_value_list = []
dummified_variables = ['city_Astapor', "city_King's Landing", 'city_Winterfell', 'phone_Android', 'phone_iPhone']

for categorical in dummified_variables:
    a = churn_df[categorical]
    b = churn_df['retained']

    data = pd.concat([a, b], axis=1)#.reset_index(drop=True)
    contingency_table = np.array([[data.groupby([a,b]).count()[categorical].unstack()[0][0], data.groupby([a,b]).count()[categorical].unstack()[1][0]], 
                                [data.groupby([a,b]).count()[categorical].unstack()[0][1], data.groupby([a,b]).count()[categorical].unstack()[1][1]]])
    chi2, p, dof, expected = sp.stats.chi2_contingency(contingency_table)
    chi2_list.append(chi2)
    p_value_list.append(p)
```

```python
p_value_cutoff = 0.05

categorical_features_p_value_dict = dict(zip(dummified_variables, p_value_list))
significant_categorical_features = [name for (name, pval) in categorical_features_p_value_dict.items() if pval < p_value_cutoff]
print len(significant_categorical_features), "categorical features has a chi-square test with the independent label with a p-value lower than %2.5f" % p_value_cutoff
```

```python
#Setting up K-fold CV function:
from sklearn.cross_validation import KFold, StratifiedKFold 
def run_cv(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    # kf = KFold(len(y),n_folds=4,shuffle=True)
    #Stretified KFold preserves the class balance in the training and test samples
    #Better for when there's a class imbalance and after up/down-sampling
    kf = StratifiedKFold(y, n_folds=5)
    
    y_pred = y.copy()
    
    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred, clf
```

### Calibration and Discrimination

```python
X = X_final_np
Y = Y_final_np
```

```python
#Setting up K-fold CV function:
from sklearn.cross_validation import KFold, StratifiedKFold 
def run_prob_cv(X,y,clf_class,**kwargs):
    kf = StratifiedKFold(y, n_folds=2)
    y_prob = np.zeros((len(y),2))
    
    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_prob[test_index] = clf.predict_proba(X_test)
    return y_prob
```

```python
def calibration(prob, outcome, n_bins=10):
    prob = np.array(prob)
    outcome = np.array(outcome)

    c = 0.0
    # Construct bins
    judgement_bins = np.arange(n_bins + 1) / n_bins
    # Which bin is each prediction in?
    bin_num = np.digitize(prob, judgement_bins)
    for j_bin in np.unique(bin_num):
        # Is event in bin
        in_bin = bin_num == j_bin
        # Predicted probability taken as average of preds in bin
        predicted_prob = np.mean(prob[in_bin])
        # How often did events in this bin actually happen?
        true_bin_prob = np.mean(outcome[in_bin])
        # Squared distance between predicted and true times num of obs
        c += np.sum(in_bin) * ((predicted_prob - true_bin_prob) ** 2)
    return c / len(prob)
```

```python
def discrimination(prob, outcome, n_bins=10):
    prob = np.array(prob)
    outcome = np.array(outcome)

    d = 0.0
    # Base frequency of outcomes
    base_prob = np.mean(outcome)
    # Construct bins
    judgement_bins = np.arange(n_bins + 1) / n_bins
    # Which bin is each prediction in?
    bin_num = np.digitize(prob, judgement_bins)
    for j_bin in np.unique(bin_num):
        in_bin = bin_num == j_bin
        true_bin_prob = np.mean(outcome[in_bin])
        # Squared distance between true and base times num of obs
        d += np.sum(in_bin) * ((true_bin_prob - base_prob) ** 2)
    return d / len(prob)
```

```python
def print_measurements(pred_prob):
    churn_prob, is_churn = pred_prob[:, 1], Y == 1
    print "  %-20s %.4f" % ("Calibration Error", calibration(churn_prob, is_churn))
    print "  %-20s %.4f" % ("Discrimination", discrimination(churn_prob, is_churn))
```

```python
from sklearn.ensemble import RandomForestClassifier as RF
print "Random forests:", print_measurements(run_prob_cv(X, Y, RF, n_estimators=100))
```

### Kaplan-Meier Survival Curve Analysis

```python
survival_df = churn_df_pre_transform
```

```python
#1 = Died/Churned, 0 = Survived/Retained
survival_df['E'] = 1 - survival_df['retained']
```

```python
#T: Total days for Retained/survived = 5 months after signup (about 5*30)
#T: Total days for Churned/died = signup_date to last_trip_date (i.e. last_trip_date is when they churned/'died')
survival_df['T'] = np.where(survival_df["E"]==1, (survival_df.last_trip_date - survival_df.signup_date).dt.days, 5*30)
```

```python
survival_df.head()
```

```python
T = survival_df['T']
E = survival_df['E']
```

```python
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
kmf.fit(T, event_observed=E, label = "Customer lifetime"); # more succiently, kmf.fit(T,E)

kmf.survival_function_
kmf.median_
kmf.plot(xlim = [0,100]);
```

```python
#King's Landing
groups = survival_df["city_King's Landing"]
King = (groups == 1)

kmf.fit(T[~King], E[~King], label="Customers NOT from King's Landing")
ax = kmf.plot()

kmf.fit(T[King], E[King], label="Customers from King's Landing")
kmf.plot(xlim=[0,100],ax=ax,title="Kaplan-Meier Survival Curves for Customers from King's Landing");
```

```python
#Trips first 30 days
groups = survival_df["trips_in_first_30_days"]
trips = (groups >= 10)

kmf.fit(T[~trips], E[~trips], label="Customers who took LESS than 5 trips during first 30 days")
ax = kmf.plot()

kmf.fit(T[trips], E[trips], label="Customers who took MORE than 5 trips during the first 30 days")
kmf.plot(xlim=[0,100],ax=ax,title="Kaplan-Meier Survival Curves for Customers who took different number of trips");
```
